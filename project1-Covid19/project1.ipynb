#Fixing the date (months to date )

```python
data['Date'] = pd.to_datetime(data['Date'], format='%b-%y')
```

```python
data.head(10)
```

# Handling Missing Values

```python
data =  data.sort_values('Date').reset_index(drop=True)
```

```python
data['Date'].head()
```

```python
data
```

```python
missing_percent = data.isnull().mean() * 100
```

```python
print(missing_percent)
```

```python
count_cols = [col for col in data.columns
              if 'Confirmed cases' in col or 'Deaths' in col]
```

```python
count_cols
```

```python
data[count_cols] = data[count_cols].ffill().bfill()
```

```python
data[count_cols].isnull().sum()
```

```python
divisions = ['Barisal', 'CTG', 'DHK', 'Khulna', 'MMsingh',
             'Rajshahi', 'Rangpur', 'Sylhet', 'allover Bangladesh']

for div in divisions:
    confirmed = f'Confirmed cases {div}'
    deaths    = f'Deaths {div}'

    data[f'Case rate {div}'] = (data[confirmed] / data[confirmed].max()) * 100
    data[f'Death Rate {div}'] = (data[deaths] / data[confirmed].replace(0, np.nan)) * 100

print("Rates recalculated successfully")
```

```python
data.isnull().sum()
```

# Data Visualization

```python
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['Confirmed cases allover Bangladesh'], 
         color='steelblue', linewidth=2, marker='o', markersize=4)

plt.title('COVID-19 Confirmed Cases Over Time — Bangladesh', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='Date', y='Deaths allover Bangladesh',
             color='crimson', linewidth=2, markers=True)

plt.title('COVID-19 Deaths Over Time — Bangladesh', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Deaths', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Confirmed cases for all divisions over time

```python
divisions = ['Barisal', 'CTG', 'DHK', 'Khulna', 'MMsingh',
             'Rajshahi', 'Rangpur', 'Sylhet']

plt.figure(figsize=(16, 7))

for div in divisions:
    plt.plot(data['Date'], data[f'Confirmed cases {div}'], 
             linewidth=2, label=div, marker='o', markersize=3)

plt.title('Confirmed Cases by Division Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)
plt.legend(loc='upper left', fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

total cases per division

```python
total_cases = {div: data[f'Confirmed cases {div}'].max()
               for div in divisions}


plt.figure(figsize=(12, 6))
sns.barplot(x=list(total_cases.keys()), y=list(total_cases.values()),
            palette='viridis')
plt.title('Peak Confirmed Cases by Division', fontsize=16)
plt.xlabel('Division', fontsize=12)
plt.ylabel('Peak Confirmed Cases', fontsize=12)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
total_deaths = {div: data[f'Deaths {div}'].sum() 
                for div in divisions}

plt.figure(figsize=(12, 6))

# Use hue to fix the deprecation warning
sns.barplot(x=list(total_deaths.keys()), 
            y=list(total_deaths.values()),
            hue=list(total_deaths.keys()),
            palette='Reds_d',
            legend=False)

plt.title('Peak Deaths by Division', fontsize=16)
plt.xlabel('Division', fontsize=12)
plt.ylabel('Peak Deaths', fontsize=12)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
